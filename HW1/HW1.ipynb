An identity matrix, often denoted as "I" or "I_n" (where "n" represents the dimension of the square matrix), is a special type of square matrix that has the following properties:

1. Diagonal Elements: All diagonal elements of an identity matrix are equal to 1.

2. Off-Diagonal Elements: All off-diagonal elements (elements that are not on the main diagonal) are equal to 0.

The identity matrix is always square, meaning it has the same number of rows and columns. It is commonly encountered in linear algebra and various mathematical operations.

Multiplying a matrix by an identity matrix is a fundamental operation in linear algebra. When you multiply any matrix "A" by an identity matrix "I," the result is the original matrix "A" itself. In other words, the identity matrix acts as a neutral element for matrix multiplication, similar to how the number 1 acts as a neutral element for multiplication of real numbers.

Mathematically, if "A" is an arbitrary matrix and "I" is the identity matrix of the appropriate size such that the multiplication is defined, then:

A x I = A

This property is analogous to the fact that multiplying any number by 1 results in the original number. The identity matrix preserves the original matrix's properties, dimensions, and values.

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

To determine if matrix T is invertible, we can check its determinant. If the determinant of T is nonzero, then T is invertible.  If the determinant of T is zero, it means that T is not invertible, and the original image cannot be retrieved using this method. In such cases, you might need to consider alternative approaches to recover the original image.

```python
import numpy as np
import matplotlib.pyplot as plt
determinant_T = np.linalg.det(T)
if determinant_T != 0:
    Tinv = np.linalg.inv(T)
    original_image = np.dot(Tinv, transformed_image)
    plt.imshow(original_image)
    plt.title("Retrieved Original Image")
    plt.show()
else:
    print("Matrix T is not invertible.")
```

## Part 8: Extra Points

* We want to perform the same operations but on the columns of the original image. How would you do the matrix algebra? Plot the scrambled image for the proof of concept.


```python
import numpy as np
import matplotlib.pyplot as plt
#Orignal image is a 3D image
num_rows, num_cols, num_channels = image.shape
col_permutation_matrix = np.eye(num_cols)
np.random.shuffle(col_permutation_matrix)
shuffled_img = np.zeros_like(image)
for channel in range(num_channels):
    shuffled_img[:, :, channel] = np.dot(image[:, :, channel], col_permutation_matrix.T)

plt.imshow(shuffled_img)  #
plt.title("Scrambled Columns of Original Image")
plt.tight_layout()
plt.show()

inverse_col_permutation_matrix = np.linalg.inv(col_permutation_matrix)
recovered_image = np.zeros_like(image)
for channel in range(num_channels):
    recovered_image[:, :, channel] = np.dot(shuffled_img[:, :, channel], inverse_col_permutation_matrix.T)


plt.imshow(recovered_image)
plt.title("Recovered Original Image")
plt.tight_layout()
plt.show()
```

Explanation -  we create and shuffle a permutation matrix (col_permutation_matrix) for the columns of each color channel separately. Then, we apply the shuffle to the original image matrix for each channel individually. Finally, we compute the inverse permutation matrix to recover the original image for each color channel and combine them to get the full color image

```python
import numpy as np
import matplotlib.pyplot as plt
num_rows, num_cols, num_channels = image.shape
row_permutation_matrix = np.eye(num_rows)
col_permutation_matrix = np.eye(num_cols)
np.random.shuffle(row_permutation_matrix)
np.random.shuffle(col_permutation_matrix)
shuffled_img = np.zeros_like(image)
for channel in range(num_channels):
    shuffled_img[:, :, channel] = np.dot(np.dot(row_permutation_matrix, image[:, :, channel]),
                                         col_permutation_matrix.T)
plt.imshow(shuffled_img)
plt.title("Scrambled Rows and Columns of Original Image")
plt.tight_layout()
plt.show()

inverse_row_permutation_matrix = np.linalg.inv(row_permutation_matrix)
inverse_col_permutation_matrix = np.linalg.inv(col_permutation_matrix)

recovered_image_new = np.zeros_like(image)
for channel in range(num_channels):
    recovered_image_new[:, :, channel] = np.dot(np.dot( inverse_row_permutation_matrix,shuffled_img[:, :, channel]),
                                             inverse_col_permutation_matrix.T)


plt.imshow(recovered_image_new)
plt.title("Recovered Original Image")
plt.tight_layout()
plt.show()
```

Explanantion -
We can first shuffle the rows and columns independently using permutation matrices and then combine the shuffling effects.
we first create and shuffle separate permutation matrices for rows and columns. Then, we apply these shufflings to the original image matrix for each color channel individually. Finally, we compute the inverse permutation matrices and apply them to recover the original image for each color channel.
